**6. How many states have both Republican senators?**

```python
states=list(data["STATE"])
party=list(data["PARTY"])
d={}

for s in states:
  d[s]=[]

def get_index(x,L):
  ls=[]
  for i in range(len(L)):
    if L[i]==x:
      ls.append(i)
  return ls

for key in d:
  m=get_index(key,states)
  d[key].append(party[m[0]])
  d[key].append(party[m[1]])

count=0
for key in d:
  if d[key][0]==d[key][1]=="R":
    count+=1
print(count)
```

**Ans. On the basis od above code the answer is 22.**
