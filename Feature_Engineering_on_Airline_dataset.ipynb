```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("/content/Data_Train.xlsx")
train_df.head()
```

```python
test_df = pd.read_excel("/content/Test_set.xlsx")
test_df.head()
# if we compare train_df and test_df, can see that Price column is missing from the test_df
```

```python
final_df = train_df.append(test_df)
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
# creating three features from the date_of_Journey column
```

```python
final_df["Day"]=final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Month"]=final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Year"]=final_df["Date_of_Journey"].str.split("/").str[2]
```

```python
final_df.head(2)
```

```python
final_df["Day"]=final_df["Day"].astype(int)
final_df["Month"]=final_df["Month"].astype(int)
final_df["Year"]=final_df["Year"].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Date_of_Journey",axis=1,inplace=True) # axis=1 for column removal and axis=0 for row removal
```

```python
final_df.head()
```

```python
final_df["Arrival_Time"]=final_df["Arrival_Time"].str.split(" ").str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df["Arrival_Hour"]=final_df["Arrival_Time"].str.split(":").str[0].astype(int)
final_df["Arrival_Minutes"]=final_df["Arrival_Time"].str.split(":").str[1].astype(int)
final_df.head()
```

```python
final_df.drop("Arrival_Time",axis=1,inplace=True)
final_df.head()
```

```python
final_df["Dep_Hour"]=final_df["Dep_Time"].str.split(":").str[0].astype(int)
final_df["Dep_Minutes"]=final_df["Dep_Time"].str.split(":").str[1].astype(int)
final_df.drop("Dep_Time",axis=1,inplace=True)
final_df.head()
```

```python
final_df.info()
```

```python
final_df[final_df["Total_Stops"].isnull()]
```

```python
final_df["Total_Stops"].unique()
```

```python
final_df["Total_Stops"]=final_df["Total_Stops"].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,'nan':1}) # nan=1 because at leat one should be there from delhi to cochin
final_df.head()
```

```python
final_df.drop("Route",axis=1,inplace=True)
final_df.head()
```

```python
final_df["duration_hour"]=final_df["Duration"].str.split(" ").str[0].str.split('h').str[0]
final_df.head()
```

```python
final_df[final_df["Duration"]=="5m"]
```

```python
final_df.drop([6474,2660],axis=0,inplace=True)
```

```python
final_df["duration_hour"]=final_df["duration_hour"].astype(int)*60   #converts the duration hour in minutes
```

```python
final_df.drop("Duration",axis=1,inplace=True)
```

```python
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df["Airline"]=labelencoder.fit_transform(final_df["Airline"])
final_df["Source"]=labelencoder.fit_transform(final_df["Source"])
final_df["Destination"]=labelencoder.fit_transform(final_df["Destination"])
final_df["Additional_Info"]=labelencoder.fit_transform(final_df["Additional_Info"])
```
